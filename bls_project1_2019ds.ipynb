```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ds_2019 = pd.read_csv(r"C:\Users\dtrob\repos\Project-1\Project-1\national_M2019_dl.csv")[["occ_title", "tot_emp", "h_mean", "a_mean"]]
print("Student data read successfully!")
```

```python
ds_2019.rename(columns={"occ_title":"Occupations", "tot_emp":"Number Employed", "h_mean":"Mean Hourly Wage", "a_mean":"Mean Annual Wage"}, inplace=True)
ds_2019.head()
```

```python
analysts = ds_2019[ds_2019["Occupations"].str.contains("Analysts$")]
analysts.head()
```

```python
x_axis = np.arange(len(analysts))
numemployed_axis = analysts["Number Employed"]
tick_locations = [value for value in x_axis]

plt.figure(figsize=(5, 4))
plt.bar(x_axis, numemployed_axis, color="r", alpha=0.5, align="center")
plt.xticks(tick_locations, analysts["Occupations"], rotation="vertical")

plt.title("Number Employed per Analyst Occupation")
plt.xlabel("Occupation")
plt.ylabel("Number Employed")

plt.show()
```

```python
meanhrwage_axis = analysts["Mean Hourly Wage"]
plt.bar(x_axis, meanhrwage_axis, color="r", alpha=0.5, align="center")
plt.xticks(tick_locations, analysts["Occupations"], rotation="vertical")

plt.title("Mean Hourly Wage per Analyst Occupation")
plt.xlabel("Occupation")
plt.ylabel("Mean Hourly Wage")

plt.show()
```

```python
meanannwage_axis = analysts["Mean Annual Wage"]
plt.bar(x_axis, meanannwage_axis, color="r", alpha=0.5, align="center")
plt.xticks(tick_locations, analysts["Occupations"], rotation="vertical")

plt.title("Mean Annual Wage per Analyst Occupation")
plt.xlabel("Occupation")
plt.ylabel("Mean Annual Wage")

plt.show()
```
